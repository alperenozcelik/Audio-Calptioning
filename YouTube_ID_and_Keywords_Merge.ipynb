```python
df = pd.read_csv('/content/drive/MyDrive/balanced_train_segments.csv',sep=', ')
df = df.drop(['start_seconds', 'end_seconds'], axis=1)
df = df.rename(columns={'"YTID': 'YTID', 'positive_labels"': 'labels'})
```

```python
df.head()
```

```python
def Convert(string):
    l = string.split(",")
    return l

temp = pd.DataFrame()

for i in df['YTID']:
  df['YTID'] = df['YTID'].replace([i],i.replace('"', ''))

x = df.YTID

for i in df['labels']:
  new_string = i.replace('"', '')
  l = Convert(new_string)
  temp = temp.append({'label': i,  'labels': l}, ignore_index=True)
  temp['YTID'] = x
  df = temp.drop(['label'], axis=1)
```

```python
df.head()
```

```python
df2 = pd.read_csv("/content/drive/MyDrive/class_labels_indices.csv")
df2 = df2.drop(['index'], axis=1)
df2.head()
```

```python
all_keywords_list  = []
for i in range(len(df['labels'])):
  keywords = ""
  for j in df['labels'][i]:
    for k in range(len(df2['mid'])):
      if j == df2['mid'][k]:
        keywords += df2['display_name'][k] + ";"
  keywords = keywords[:-1]
  all_keywords_list.append(keywords)
```

```python
df['Keywords'] = all_keywords_list
df = df.drop(['labels'], axis=1)
df
```

```python
df.to_csv ("YouTubeID_and_Keywords.csv", index=False, header=True)
```
